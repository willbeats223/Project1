```python
# Import the required libraries and dependencies
import datetime as dt
import pandas as pd
from pandas_datareader import data as pdr
import yfinance as yf
import hvplot
import hvplot.pandas
```

```python
# Using the read_csv function and Path module, create a DataFrame 
end = dt.datetime.now()
start = dt.datetime(2008,1,1)
start, end
```

```python
tickers = ["SUSA", "XLY", "XLP", "XLE", "XME", "VIS", "XLV", "XLF", "VGT", "VOX", "VPU", "IYR", "SPY", "QQQ"]
df_closing_price = yf.download(tickers, period = '15y',interval='1wk')['Adj Close']
df_closing_price.head()
#data = df_closing_price
```

```python
df_closing_price = df_closing_price.reset_index()
df_closing_price['Date'] = pd.to_datetime(df_closing_price['Date'])
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'
df_closing_price['season'] = df_closing_price['Date'].dt.month.apply(get_season)
display(df_closing_price.head())
display(df_closing_price.tail())
```

```python
df_closing_price = df_closing_price.set_index('Date')
```

```python
def plot_year(year):
    df_year = df_closing_price[df_closing_price.index.year == year]
    return df_year.hvplot.line()

from ipywidgets import interact

@interact(year=(df_closing_price.index.year.min(),df_closing_price.index.year.max()))
def plot_selected_year(year):
    return plot_year(year)
```

```python
def plot_season(season):
    df_season = df_closing_price[df_closing_price['season'] == season]
    return df_season.hvplot.line()
from ipywidgets import interact
@interact(season=df_closing_price['season'].unique())
def plot_selected_season(season):
    return plot_season(season)
```
